```python
#Matthew Stewart
#Assignment 14

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Cannot divide by zero"
    return x / y

def calculator():
    while True:
        user_input = input("Enter a problem (Example: '13 / 5'): ")

        if user_input.lower() == "exit":
            break

        parts = user_input.split()
        if len(parts) == 3:
            num1 = float(parts[0])
            operator = parts[1]
            num2 = float(parts[2])

            if operator == "+":
                print("Result: " + str(add(num1, num2)))
            elif operator == "-":
                print("Result: " + str(subtract(num1, num2)))
            elif operator == "*":
                print("Result: " + str(multiply(num1, num2)))
            elif operator == "/":
                print("Result: " + str(divide(num1, num2)))
            else:
                print("Invalid operator. Please use +, -, *, or /.")
        else:
            print("Invalid input format. Please use 'number operator number' (Example: '13 / 5').")

calculator()

#Assignment 13 Enhanced
import random

import random

class Person:
    def __init__(self):
        self.first_name = self.gen_first_name()
        self.last_name = self.gen_last_name()
        self.email = self.gen_default_email(self.first_name, self.last_name)
        self.phone_number = self.gen_phone_number()
        self.pastimes = self.gen_pastimes()

    def gen_first_name(self):
        first_names = ['Matt', 'Ava', 'Zackson', 'David', 'Tom']
        return random.choice(first_names)

    def gen_last_name(self):
        last_names = ['Stewart', 'Johnson', 'kaufman', 'Yash', 'Thomas']
        return random.choice(last_names)

    def gen_default_email(self, first_name, last_name):
        return f"{first_name.lower()}.{last_name.lower()}@example.com"

    def gen_phone_number(self):
        return f"+1 {random.randint(100, 999)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}"

    def gen_pastimes(self):
        pastimes = ['skydiving', 'piano', 'football', 'coding', 'swimming']
        return ', '.join(random.sample(pastimes, random.randint(1, len(pastimes))))

    def as_transmitted(self):
        return f"{self.first_name} {self.last_name}, {self.email}, {self.phone_number}, {self.pastimes}"

    def as_received(self):
        received_data = self.as_transmitted()
        return received_data.replace(' ', '_')

    def __str__(self):
        return self.as_transmitted()

    def __lt__(self, other):
        return self.first_name < other.first_name


def write_file(file, data):
    try:
        with open(file, 'w') as f:
            f.write(data)
    except Exception as e:
        print(f"Error writing to {file}: {str(e)}")

def make_files(my_name, record_count=100):
    random.seed(hash(my_name))

    people = [Person() for _ in range(record_count)]

    source_file = 'source.txt'
    received_file1 = 'as_received1.txt'
    received_file2 = 'as_received2.txt'
    received_file3 = 'as_received3.txt'

    all_files = [source_file, received_file1, received_file2, received_file3]

    try:
        with open(source_file, 'w') as src_f, \
             open(received_file1, 'w') as rec1_f, \
             open(received_file2, 'w') as rec2_f, \
             open(received_file3, 'w') as rec3_f:

            header = 'Name, Email, Phone, Pastimes\n'
            src_f.write(header)
            rec1_f.write(header)
            rec2_f.write(header)
            rec3_f.write(header)

            source_file_line_count = 0
            received_file1_line_count = 0
            received_file2_line_count = 0
            received_file3_line_count = 0

            for person in sorted(people):
                source_data = person.as_transmitted() + '\n'
                received_data = person.as_received() + '\n'

                src_f.write(source_data)
                source_file_line_count += 1

                choice = random.choice([1, 2, 4, 1, 2, 4, 1, 2, 4, 1, 2, 4, 3, 5, 6])
                if choice & 1:
                    rec1_f.write(received_data)
                    received_file1_line_count += 1
                if choice & 2:
                    rec2_f.write(received_data)
                    received_file2_line_count += 1
                if choice & 4:
                    rec3_f.write(received_data)
                    received_file3_line_count += 1

            print(f'source_file written with {source_file_line_count} lines')
            print(f'received_file1 written with {received_file1_line_count} lines')
            print(f'received_file2 written with {received_file2_line_count} lines')
            print(f'received_file3 written with {received_file3_line_count} lines')

    except Exception as e:
        print(f"An error occurred: {str(e)}")

make_files(my_name='Matt Stewart', record_count=20)
```

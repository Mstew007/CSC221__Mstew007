```python
# Assignment 9
# Matthew Stewart

#8.2
def favorite_book(title):
    print(title + " is my favorite book.")

favorite_book('Harry Potter and the sorcerers stone')

#8.3
def make_shirt(size, message):
    print("\nI am going to make a " + size + " t-shirt.")
    print('the shirt will say, "' + message + '"')

make_shirt('medium', 'I love Lambos!')
make_shirt(message="Python>snakes.", size='medium')

#8.4
def make_shirt(size='large', message='I love Python!'):
    print("\nI am going to make a " + size + " t-shirt.")
    print('It will say, "' + message + '"')

make_shirt()
make_shirt(size='medium')
make_shirt('small', 'I hate snakes.')

#8.14
def make_car(manufacturer, model, **options):
    car_dict = {
        'manufacturer': manufacturer.title(),
        'model': model.title(),
        }
    for option, value in options.items():
        car_dict[option] = value

    return car_dict

my_bmw = make_car('bmw', '325i', color='black', tow_package=True)
print(my_bmw)

my_audi = make_car('audi', 'rs3', year=2022, color='silver',
        headlights='popup')
print(my_audi)

# number 2
import math

def two_numbers(num1, num2=math.pi):
    total = num1 + num2

    sorted_list = sorted([num1, num2])

    return total, sorted_list

result1 = two_numbers(6, 9)
print("Sum:", result1[0])
print("Sorted Numbers:", result1[1])

result2 = two_numbers(7)
print("Sum:", result2[0])
print("Sorted Numbers:", result2[1])
print("\n")

#number 3
result1 = two_numbers(4, 5)
print("Test #1:", result1)

result2 = two_numbers(2)
print("Test #2:", result2)

result3 = two_numbers(1, 5)
print("Test #3:", result3)

result4 = two_numbers(-5, -1)
print("Test #4:", result4)

result5 = two_numbers(1000000, 5000000)
print("Test #5:", result5)
print("\n")

#number 4
import math

def two_numbers(num1, num2=math.pi, verbose=False):
    total = num1 + num2

    sorted_list = sorted([num1, num2])

    if verbose:
        print("Parameters: num1 =", num1, "num2 =", num2)
        print("Sum:", total)
        print("Sorted Numbers:", sorted_list)

    return total, sorted_list

result1 = two_numbers(6, 9, verbose=True)
result2 = two_numbers(8, verbose=True)
```
